All imports, must be run first

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

Rainfall data set upload without use of Kmeans

```python
df=pd.read_csv('../Rainfall(edited final).csv')

# rainfall is classified as <1000 is light rain for the year, <1000 && >1300 is average and >1300 is heavy rain

#dataset size
print(f'The Rain table has {df.shape[0]} rows, and {df.shape[1]} columns')
df.iloc[:5, list(range(5)) + [-1]]
```

Rainfall dataset upload with use of clustering by KMeans

```python
df=pd.read_csv('../Rainfall(Kmeanstest).csv')

kmeans = KMeans(n_clusters=3, random_state=0).fit(df)

kmeans.labels_

df['ann'] = kmeans.predict(df)

print(f'The Rain table has {df.shape[0]} rows, and {df.shape[1]} columns')
df.iloc[:5, list(range(5)) + [-1]]
```

checking the number of entries under each label

```python
df['ann'].value_counts()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X = np.array(X)

y_labels = y
le = LabelEncoder()
le.fit(y)
y_encoded = le.transform(y)
# show few class labels after encoding
y_encoded[:10]


inversed_labels = le.inverse_transform(y_encoded)
# show few class labels after dencoding
inversed_labels[:10]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size = 0.3,random_state=30)
X_train = X_train/255.0
X_test = X_test / 255.0

# create an instance of RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,)

# fit the model on training data
rf.fit(X_train,y_train)
```

```python
# Predicted classes of the testing set
rf_predictions = rf.predict(X_test)

rf_probs = rf.predict_proba(X_test)

rf_probs[:1]

accuracy = accuracy_score(y_test, rf_predictions)
print(f'The overall accuracy of RF is {np.round(accuracy*100,2)}%')
```

```python
print(confusion_matrix(y_test,rf_predictions))
print(classification_report(y_test,rf_predictions))
```
